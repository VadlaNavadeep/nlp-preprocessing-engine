# NLP Preprocessing Engine
### Internship Task Submission

This notebook implements a robust NLP preprocessing pipeline including text cleaning, tokenization, stopword removal, and lemmatization.

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

## Preprocessing Functions

```python
def to_lower(text):
    return text.lower()

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

def remove_numbers(text):
    return re.sub(r'\d+', '', text)

def remove_urls(text):
    return re.sub(r'http\S+', '', text)

def remove_emojis(text):
    return re.sub(r'[^\x00-\x7F]+', '', text)

def tokenize(text):
    return word_tokenize(text)

stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

lemmatizer = WordNetLemmatizer()

def lemmatize(tokens):
    return [lemmatizer.lemmatize(word) for word in tokens]
```

## Final Pipeline

```python
def preprocess_text(text):
    text = to_lower(text)
    text = remove_urls(text)
    text = remove_emojis(text)
    text = remove_punctuation(text)
    text = remove_numbers(text)
    tokens = tokenize(text)
    tokens = remove_stopwords(tokens)
    tokens = lemmatize(tokens)
    return tokens
```

## Testing

```python
sample_text = "Hello!!! This is Navadeep's NLP preprocessing engine 😎. Visit http://example.com"
output = preprocess_text(sample_text)
print(output)
```
